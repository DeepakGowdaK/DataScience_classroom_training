```python
from scipy.stats import norm
x = np.arange(-3, 3, 0.01)

plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5)) 
plt.plot(x, norm.pdf(x, -1.0, 0.4) )
plt.show()
```
