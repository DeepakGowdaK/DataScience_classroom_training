#### Encode categorical features to number

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in df.columns:
    df[col] = le.fit_transform( df[col] )
    
df.head(2)
```

```python
# Split dataframe in x & y

y = df.iloc[:, 0].values    # df.iloc[ row_start:row_end, column_strt:column_end]    ]
type(y)
```

```python
x = df.iloc[:, 1:23].values
type(x)
```

```python
# Build model
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0)
```

```python
# Training the model
clf.fit(x_train, y_train)
```

```python
# Predict y_predict from x_test

y_predict = clf.predict(x_test)
```

```python
# Validation - Confusion matrics

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
cm
```

# Decision Tree Model 

### Hiring Decision Making

```python
df = pd.read_csv('PastHires.csv')
df.head(2)
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Mapping of strings (labels) to numbers
d = {'Y': 1, 'N': 0}
df['Hired'] = df['Hired'].map(d)
df['Employed?'] = df['Employed?'].map(d)
df['Top-tier school'] = df['Top-tier school']map(d)
df['Interned'] = df['Interned'].map(d).
d = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Level of Education'] = df['Level of Education'].map(d)
df.head()
```

```python
# Split data in x & y
y = df['Hired']
x = df.drop( ['Hired'], axis=1 )
```

```python
# Build model

from sklearn import tree

clf = tree.DecisionTreeClassifier()
```

```python
# Train the model
clf = clf.fit(x, y)
```

```python
# Predict for following
# exp 10 yrs, employed, 4 employers, BS, No top tier school,  no internship
clf.predict( [[10, 1, 4, 0, 0, 0]] )
```

```python
# fresher, not employed, 0 , BS, No top tier school,  no internship
clf.predict( [[0, 0, 0, 0, 0, 0]] )
```
