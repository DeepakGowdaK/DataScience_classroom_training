## **Building model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x = data.drop(['charges'], axis=1) #leaving cllumn Charges for prediction.
x.head(2)
```

```python
x.shape
```

```python
y = data.charges # considering only charges cloumn only
y.head(2)
```

```python
y.shape
```

```python
# dividing x & y in training & testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

# Put 75% data as training data
# Put 25% data as testing data

# For training of model, we will use x_train => y_train.
# For validation, we will use x_test => y_predict <--> y_test
```

```python
# Building Linear Regression model
lr = LinearRegression

# Training the model
lr = lr().fit(x_train, y_train)
lr
```

```python
# Validate model
lr.score(x_test, y_test)
```

### ***Score 1 is best, 0 is bad.***

```python
# Prediction of medical charges for
# age 18, male, bmi 24, children 0, non-smoking, region.
lr.predict([[18, 1,24,0, 0, 1]])
```

```python
# Prediction of medical charges for 
# age 80, male, bmi 60, children 4, smoking, region
lr.predict( [[80, 1, 60, 4, 1, 1]] )
```
