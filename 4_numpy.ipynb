## **Compare the performance of Python list and numpy array**

```python
import timeit
import random

# Create python list by random sampling in the range 0-999. 1000 items in list
python_list = random.sample( range(1000), 1000)
python_list
print(python_list)
print(len(python_list))
```

```python
# Create numpy array by random sampling in the range 0-999. 1000 items in list
np_array = np.random.randint( 1000, size=1000)
np_array

np_array.size
```

```python
def python_code():
    # return list where 32 will be subtracted from every element
    return [val-32 for val in python_list]

def numpy_code():
    return np_array-32
```

```python
n = 1000
t_python = timeit.timeit(python_code, number=n)
t_numpy = timeit.timeit(numpy_code, number=n)

print('Time for python : ', t_python)
print('Time for numpy : ', t_numpy)
print('Speed Comparison : ', t_python/t_numpy)
```

```python
arrangeArray = np.arange(10)
print(arrangeArray)
arrangeArray[2:8]   # get slice 
arrangeArray[::-1]   # reverse the array
arrangeArray[:3] = 100  # change value of first three elements
print(arrangeArray)
arrangeArray[:] = 100
print(arrangeArray)
```

## **Image procesing using numpy array**

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline   

pic = Image.open('puppy.jpg')
pic
```

```python
pic_array = np.asarray(pic)
print(pic_array.shape)
pic_red = pic_array.copy()
pic_green = pic_array.copy()
pic_blue = pic_array.copy()
```

```python
pic_red[:, :, 1] = 0 #green color
pic_red[:, :, 2] = 0 #blue color
plt.imshow(pic_red)
```

```python
pic_green[:, :, 0] = 0 #red color
##pic_green[:, :, 1] = 0 #green color
pic_green[:, :, 2] = 0 #blue color
plt.imshow(pic_green)
```

```python
pic_blue[:, :, 0] = 0 #red color
pic_blue[:, :, 1] = 0 #green color
##pic_blue[:, :, 2] = 0 #blue color
plt.imshow(pic_blue)
```
