# Naive Bayes Model

```python
import os
import io
import numpy
from pandas import DataFrame
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

def readFiles(path):
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            path = os.path.join(root, filename)

            inBody = False
            lines = []
            f = io.open(path, 'r', encoding='latin1')
            for line in f:
                if inBody:
                    lines.append(line)
                elif line == '\n':
                    inBody = True
            f.close()
            message = '\n'.join(lines)
            yield path, message


def dataFrameFromDirectory(path, classification):
    rows = []
    index = []
    for filename, message in readFiles(path):
        rows.append({'message': message, 'class': classification})
        index.append(filename)

    return DataFrame(rows, index=index)
```

```python
data = DataFrame({'message': [], 'class': []})
```

```python
data.head()
```

```python
data = data.append(dataFrameFromDirectory('emails/spam', 'spam'))
data = data.append(dataFrameFromDirectory('emails/ham', 'ham'))
```

```python
data.head(2)
```

```python
data.tail(2)
```

```python
data.shape
```

```python
data['class'].value_counts()
```

```python
vectorizer = CountVectorizer()
counts = vectorizer.fit_transform( data['message'].values)
```

```python
counts
```

```python
# Build Naive Bayes model
clasifier = MultinomialNB()

# Train the Naive Bayes model
targets = data['class'].values
clasifier.fit(counts, targets)
```

```python
# Predictions

example = ['Free Viagra now!!!', 'Hello Dilip, today is workding day in Pune']
example_counts = vectorizer.transform(example)
predictions = clasifier.predict(example_counts)
predictions
```

# How to identify the number of clusters ?

```python
df = pd.read_csv('wine.csv')
df.head(2)
```

```python
df.shape
```

```python
df.Wine.unique()
```

```python
df.Wine.value_counts()
```

```python
x = df.iloc[:, 1:].values   # from 1st column till last column are our input parameters
y = df.iloc[:, 0].values   # first column is wine quality information
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 11):  # no of clusters 1 to 11
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) #within cluster sum of squares
    # inertial - within cluster, sum of squares
    # wess = within cluster sum of squares
    
plt.plot(range(1,11), wcss)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('wcss')
```
