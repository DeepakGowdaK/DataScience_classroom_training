```python
df = pd.DataFrame({'id': ['a', 'b', np.nan, 'd', 'e'],'val1':[4,np.nan,7,9,5], 'val2': [8,3,np.nan, 4,np.nan]})

df 

# NaN = Not a number, C, C++ => NULL
```

```python
df.isnull() # Check for missing values
```

```python
df.val1.isnull()
```

```python
df.isnull().sum()
```

```python
# drop the rows which are containing missing values
df.dropna(axis=0)   # axis=0 for rows, axis=1 for columns
```

```python
# drop the columns which are containing missing values
df.dropna(axis=1)   # axis=0 for rows, axis=1 for columns
```

## **Fill the missing values**

```python
df
```

```python
df.fillna(0) #fill missing values by 0
```

```python
dfMandya = df.fillna('Mandya') #fill missing value by Mandya
# dfMandya = df.fillna('Mandya') #fill missing value by Mandya
dfMandya
```

```python
df # from above steps the Mandya value was made only visualization purpose.
```

```python
df.fillna(method='bfill') #fill the missing value from front value.
```

```python
df.fillna(method='ffill') #fill the missing value from front value.
```

```python
df.fillna(df.mean()) # fill missing values by mean of column
```

```python
df.fillna(df.mode()) #fill missing values by most repeated value
```

```python
df.fillna(df.max()) #fill missing values by maximum value
```

```python
df.fillna(df.min()) #fill missing values by minimum value
```
